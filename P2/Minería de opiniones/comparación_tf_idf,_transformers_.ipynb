```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay, confusion_matrix
import torch
from umap import UMAP
from sklearn.preprocessing import MinMaxScaler
from datasets import DatasetDict, Dataset,ClassLabel
import matplotlib.pyplot as plt
from transformers import AutoTokenizer, DistilBertTokenizer, DistilBertModel, AutoModel, TFAutoModel, AutoModelForSequenceClassification, Trainer, TrainingArguments
from huggingface_hub import notebook_login
```

## Lectura y procesamiento de datos:

```python
notebook_login()
```

```python
data = pd.read_csv("amazon_baby.csv") # Lectura de los datos
data.head()
```

```python
data = data.dropna()
```

```python
dat = data.loc[data["rating"] != 3] # Ignoramos los valores 3 de la columna rating
dat["polaridad"] = dat["rating"].apply(lambda x: 1 if x > 3 else 0) # En la columna polaridad indicamos que si el rating es mayor a 3 le asigne el valor de 1 en caso contrario que se asigne -1
len(dat)
```

```python
dat = dat.sample(n = 10000)
dat.head()
```

```python
((dat["polaridad"]+1)/2).mean()
```

## Selecciona aleatoriamente 10000 (muestreo estratificado) comentarios y divídelos en conjuntos de entrenamiento y prueba (proporción 80-20) .

```python
X_train, X_test, y_train, y_test = train_test_split(dat["review"],dat["polaridad"], stratify=dat["polaridad"], test_size=0.2)
```

## Transforma los comentarios con la transformación tf-idf y construye un clasificador utilizando regresión logística.

```python
vectorizer = TfidfVectorizer(stop_words="english", max_features=100)
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.fit_transform(X_test)
```

### Aplicacion de regresión logística a los comentarios trasformados por tf-idf

```python
lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, y_train)
```

### Evaluacion de los resultados de tf-idf con regresión logística:
Podemos observar que la exactitud es del 78% con regresion logistica y tf-idf


```python
y_pred = lr.predict(X_test)
resTfidf = accuracy_score(y_test, y_pred)*100

print('Accuracy:', resTfidf ,'%')
```

## Vectoriza los comentarios mediante los transformers

```python
dat.head()
```

```python
data_set= pd.DataFrame({"text": dat["review"], "label": dat["polaridad"]})
```

```python
data_set.head()
```

```python
X_train, X_test = train_test_split(data_set, test_size=0.2, stratify=dat["polaridad"])
```

Se crea los conjuntos de entrenamiento y prueba a partir.

```python
train_dataset = Dataset.from_pandas(X_train)
test_dataset = Dataset.from_pandas(X_test)
```

Se elimina la columna index generada auntomaticamente

```python
train_dataset = train_dataset.remove_columns('__index_level_0__')
test_dataset = test_dataset.remove_columns('__index_level_0__')
```

Se utiliza la función "DatasetDict()"  que se utiliza para crear un diccionario para trabajar con conjuntos de datos.

```python
dataset_dict = DatasetDict({"train": train_dataset, "test": test_dataset})
dataset_dict["train"].features
```

```python
dataset_dict.set_format(type="pandas")
df_pandas = dataset_dict["train"][:]
df_pandas.head()
```

### Tokenizacion de los datos

Restablecemos el formato del conjunto de datos en el diccionario "dataset_dict" al formato interno de la biblioteca "datasets", para evitar inconvenientes:

```python
dataset_dict.reset_format()
```

Se seleccionara el modelo "distilbert-base-uncased" para proceder a la tokenizacion

```python
model_ckpt = "distilbert-base-uncased"
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AutoModel.from_pretrained(model_ckpt).to(device)
```

```python
def tokenize(batch):
    return tokenizer(batch["text"], padding=True, truncation=True, max_length=100)
```

```python
dataset_encoded = dataset_dict.map(tokenize, batched=True,batch_size=None)
```

```python
def extract_hidden_states(batch):
    inputs  = {k:v.to(device) for k,v in batch.items() if k in tokenizer.model_input_names}
    with torch.no_grad():
        last_hidden_state = model(**inputs).last_hidden_state
    return {"hidden_state": last_hidden_state[:,0].cpu().numpy()} 
```

Se establece el formato de los datos para hacer uso de tensor de PyTorch.


```python
dataset_encoded.set_format("torch", columns=["input_ids", "attention_mask", "label"])
```

```python
dataset_hidden = dataset_encoded.map(extract_hidden_states, batched=True, batch_size=32) 
```

Transformamos los datos de entrenamiento y prueba en vectores utilizando DistilBERT, los transformamos en matrices de NumPy y los separamos en grupos de características y etiquetas de clase para su posterior implementación en el entrenamiento y la evaluación del modelo de regresión logística.

```python
X_train = np.array(dataset_hidden["train"]["hidden_state"])
X_test = np.array(dataset_hidden["test"]["hidden_state"])
y_train = np.array(dataset_hidden["train"]["label"])
y_test = np.array(dataset_hidden["test"]["label"])

X_train.shape, X_test.shape
```

## Construir un clasificador basado en regresión logística (exactamente igual al punto anterior, la única diferencia debe ser el proceso de vectorización). Utiliza el modelo "distilbert-base-uncased"

Aplicamos la regresión logística para los **transformers**.

```python
lr_clf = LogisticRegression(max_iter = 1000)
lr_clf.fit(X_train, y_train)
```

Observamos que la exactitud con este modelo es de un 91%

```python
y_pred = lr_clf.predict(X_test)
resTrLr = accuracy_score(y_test, y_pred)*100 
print('Accuracy:', resTrLr ,'%')
```

## Basándote en el modelo preentrenado "distilbert-base-uncased", construye un clasificador para los comentarios de amazon haciendo un ajuste fino del modelo (fine tuning)

Se crea una instacia para el modelo:

```python
num_labels = 2
model = (AutoModelForSequenceClassification.from_pretrained(model_ckpt, num_labels=num_labels).to(device))
```

```python
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    f1 = f1_score(labels, preds, average="weighted")*100
    acc = accuracy_score(labels, preds)*100
    recall = recall_score(labels, preds, average="weighted")*100
    prec = precision_score(labels, preds, average="weighted")*100
    return {"Accuracy": acc, "Precision": prec, "Recall": recall, "F1 score": f1}
```

```python
batch_size = 32
logging_steps = len(dataset_encoded["train"]) // batch_size
model_name = f"{model_ckpt}-finetuned-emotion"
training_args = TrainingArguments(output_dir=model_name,
                                  num_train_epochs=2,
                                  learning_rate=2e-5,
                                  per_device_train_batch_size=batch_size,
                                  per_device_eval_batch_size=batch_size,
                                  weight_decay=0.01,
                                  evaluation_strategy="epoch",
                                  disable_tqdm=False,
                                  logging_steps=logging_steps,
                                  push_to_hub=True, 
                                  log_level="error")
```

Es obtiene el modelo pre entrenado.

```python
trainer = Trainer(model=model, args=training_args, 
                  compute_metrics=compute_metrics,
                  train_dataset=dataset_encoded["train"],
                  eval_dataset=dataset_encoded["test"],
                  tokenizer=tokenizer)
trainer.train();
```

```python
preds_output = trainer.predict(dataset_encoded["test"])
```

Finalmente obtenemos el resultado de nuestro clasificado dandonos la exactitud mas alta de todos el clasificadores con un total de 93%.

```python
labels = preds_output.label_ids
y_preds = np.argmax(preds_output.predictions, axis=1)
resTuring = accuracy_score(labels, y_preds)*100
print('Accuracy:', resTuring ,'%')
```
